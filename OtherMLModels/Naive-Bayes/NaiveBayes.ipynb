```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
data = load_breast_cancer()
X = data.data
Y = data.target
```

```python
data
```

```python
X_df = pd.DataFrame(X)
Y_df = pd.DataFrame(Y)
```

```python
X_df.describe()
```

```python
Y_df.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_df, Y_df, test_size = 0.20, random_state = 21)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB
def gausian_naive_base(X_tain_val, y_train_val, X_test_val):
    gnb = GaussianNB()
    gnb.fit(X_tain_val, y_train_val)
    return gnb.predict(X_test_val)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
def calculate_scores(y_test_val, y_pred_val):   
    acc_score = accuracy_score(y_test_val, y_pred_val)
    conf_matrix = confusion_matrix(y_test_val, y_pred_val)
    print("Accuracy Score:", acc_score)
    print("Confusion Matrix:", conf_matrix)
```

```python
print("Calculating Gausion Naive Bayes without any noise")
y_pred_without_noise = gausian_naive_base(X_train, y_train.values.ravel(), X_test)
calculate_scores(y_test, y_pred_without_noise)
```

```python
def calculate_gausian_naive_base_with_noise(X_train_val, y_train_val, X_test_val, y_test_val):
    y_pred_with_noise = gausian_naive_base(X_train_val, y_train_val, X_test_val)
    calculate_scores(y_test_val, y_pred_with_noise)    
```

```python
noise_variances = [50, 1000, 20000, 400000, 8000000, 100000000]
for noise in noise_variances:
    print("Calculating Gausion Naive Bayes with noise variance:", noise)
    X_noise = X + np.random.normal(0, np.sqrt(noise), X.shape)
    X_train_noise, X_test_noise, y_train_noise, y_test_noise = train_test_split(X_noise, Y, test_size = 0.20, random_state = 21)    
    calculate_gausian_naive_base_with_noise(X_train_noise, y_train_noise, X_test_noise, y_test_noise)
```
